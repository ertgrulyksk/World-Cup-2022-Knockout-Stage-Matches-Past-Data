```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('international_matches.csv')
display(data[0:3])
print(len(data))
print(data.columns.tolist())
```

```python
teams=['Netherlands','USA','Argentina','Australia','France','Poland','England','Senegal','Japan','Croatia','Brazil','South Korea','Morocco','Spain','Portugal','Switzerland']
teams16=[['Netherlands','USA'],['Argentina','Australia'],['France','Poland'],['England','Senegal'],['Japan','Croatia'],['Brazil','South Korea'],['Morocco','Spain'],['Portugal','Switzerland']]
teams8=[['Netherlands','Argentina'],['France','England'],['Croatia','Brazil'],['Morocco','Portugal']]
teams4=[['Argentina','Croatia'],['France','Morocco']]
teams2=[['Argentina','France'],['Croatia','Morocco']]
```

```python
selects=[]
for i in range(len(data)):
    if data.loc[i,'home_team'] in teams and data.loc[i,'away_team'] in teams:
        selects+=[i]
print(selects)
```

```python
data2=data.iloc[selects].reset_index(drop=True)
data2=data2.iloc[:,[0,1,2,9,10,11]]
display(data2)
print(data2.columns.tolist())
col0=data2.columns.tolist()
```

```python
# Final & Third-Place Playoff Past Match Results
for i in range(2):
    T=teams2[i]    
    print(T)
    datai=pd.DataFrame(columns=col0)
    for j in range(len(data2)):
        if set(data2.iloc[j,1:3])==set(T):
            datai=pd.concat([datai,data2.iloc[j:j+1]],axis=0)
    if len(datai)>0:
        datai=datai.sort_values('date',ascending=True)
    display(datai)
    print()
```
